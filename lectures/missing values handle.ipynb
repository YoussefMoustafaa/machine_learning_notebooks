```python
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})

df.fillna(value='100')
```

```python
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})

df['A'].fillna(value=df['A'].mean(),inplace=True)

#df['A'] = df['A'].fillna(value=df['A'].mean())

df['B'].fillna(value=df['B'].median(),inplace=True)
df
```

# impute data for missing values

```python
from sklearn.impute import SimpleImputer
```

```python
miles  = pd.DataFrame( {'miles runs ':[100,200,np.nan,56,62,np.nan,120,230,np.nan,400,210,np.nan,np.nan],
                        'playes id'  : [101,102,103,104,np.nan,105,107,108,109,110,112,np.nan,np.nan]                       
                       }
                     )
miles
```

```python
miles.isna().sum()
```

```python
imp_mean = SimpleImputer(strategy='mean')

imp_mean.fit_transform(miles)
```

```python
imp_median = SimpleImputer(strategy='median')

imp_median.fit_transform(miles)
```

```python
imp_mode = SimpleImputer(strategy='most_frequent')

imp_mode.fit_transform(miles)
```

```python
imp_constant = SimpleImputer(strategy='constant',fill_value = 0) # fillna(0)

imp_constant.fit_transform(miles)
```

```python
miles
```

# make column transformations

```python
from sklearn.compose import make_column_transformer
```

```python
ct = make_column_transformer(

    (imp_median,['miles runs ', 'playes id']),

    remainder='drop'
)

ct.set_output(transform='pandas')
```

```python
df_updated = ct.fit_transform(miles)
df_updated
```

```python
ct2 = make_column_transformer(

    (imp_mean,['miles runs ', 'playes id']),
    (imp_median,['miles runs ', 'playes id']),
    (imp_mode,['miles runs ', 'playes id']),
    (imp_constant,['miles runs ', 'playes id']),
    remainder='drop'
)

ct2.set_output(transform='pandas')
```

```python
df_updated2 = ct2.fit_transform(miles)
df_updated2
```
