```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import OneHotEncoder , OrdinalEncoder

from sklearn.compose import make_column_transformer
```

```python
df = pd.read_csv('50_Startups.csv')
df.head()
```

```python
df.info()
```

```python
df['State'].unique()
```

```python
df['State'].nunique()
```

# 1- make column transformer

```python
ohe = OneHotEncoder(sparse_output=False)
```

```python
ode = OrdinalEncoder()
```

```python
ct = make_column_transformer(

    (ohe, ['State']),
    (ode, ['State']),
    remainder='passthrough'
    
    )

ct.set_output(transform='pandas')
```

```python
df_new = ct.fit_transform(df)
df_new.head()
```

```python
df.head()
```

```python
ct2 = make_column_transformer(

    (ohe, ['State']),
    (ode, ['State']),
    remainder='drop'
    
    )

ct2.set_output(transform='pandas')
```

```python
df_new_2 = ct2.fit_transform(df)
df_new_2.head()
```

```python
ct3 = make_column_transformer(

    (ohe, ['State']),
    (ode, ['State']),
    ('passthrough', [ 'R&D Spend','Administration' ,'Marketing Spend', 'State','Profit']),
    remainder='drop'
    
    )

ct3.set_output(transform='pandas')
```

```python
df.columns
```

```python
df_new_3 = ct3.fit_transform(df)
df_new_3.head()
```

# 2 - ordinal encoder 

```python
states = list(df['State'].unique())
states
```

```python
encoder  = OrdinalEncoder(categories = [states])

encoder.fit_transform(df[['State']])

df.head()
```

```python
encoder.fit_transform(df[['State']])
```

```python
df['Encoded_State'] = encoder.fit_transform(df[['State']])

#df['state'] = encoder.fit_transform(df[['State']])
```

# 3- one hot encoding

```python
df.State.unique()
```

```python
df.State.value_counts()
```

```python
encoder_hot = OneHotEncoder(handle_unknown= 'ignore',sparse_output= False).set_output(transform='pandas')

state_encoder_hot = encoder_hot.fit_transform(df[['State']])

state_encoder_hot
```

```python
df = pd.concat([df, state_encoder_hot], axis=1)
df.head()
```
